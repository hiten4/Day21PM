# 1. Create synthetic datasets

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)


X_reg = np.linspace(0, 10, 100)
y_reg = 3 * X_reg + 5 + np.random.normal(0, 2, 100)


X_clf = np.linspace(0, 10, 100)
y_clf = np.where(X_clf > 5, 1, 0)
```

# 2. Train simple models

```python

coef = np.polyfit(X_reg, y_reg, 1)
y_pred_reg = coef[0] * X_reg + coef[1]


threshold = 5
y_pred_clf = np.where(X_clf > threshold, 1, 0)
```

# 3. Visualization

```python


# Regression plot
plt.figure()
plt.scatter(X_reg, y_reg, label="Actual")
plt.plot(X_reg, y_pred_reg, color="red", label="Prediction")
plt.title("Regression")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()

# Classification plot
plt.figure()
plt.scatter(X_clf, y_clf, label="Actual")
plt.scatter(X_clf, y_pred_clf, marker='x', label="Prediction")
plt.title("Classification")
plt.xlabel("X")
plt.ylabel("Class")
plt.legend()
plt.show()


print("Regression → continuous output")
print("Classification → categorical/binary output")
```

# 4. Manual regression (MSE)

```python


def mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

mse_value = mse(y_reg, y_pred_reg)
print("MSE:", mse_value)
```

# 5. Simple classification accuracy

```python
def accuracy(y_true, y_pred):
    correct = np.sum(y_true == y_pred)
    return correct / len(y_true)

acc = accuracy(y_clf, y_pred_clf)
print("Accuracy:", acc)
```

# 6. Comparison

```python
print("\nComparison:")
print("Regression → Predicts continuous values, evaluated using MSE")
print("Classification → Predicts categories, evaluated using Accuracy")
```
