```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

X = np.linspace(0, 10, 100)
y = np.sin(X) + np.random.normal(0, 0.2, 100)
```

```python
degrees = [1, 2, 5]
train_errors = []

plt.figure()

for d in degrees:
    coeffs = np.polyfit(X, y, d)
    y_pred = np.polyval(coeffs, X)

    error = np.mean((y - y_pred) ** 2)
    train_errors.append(error)

    plt.plot(X, y_pred, label=f"Degree {d}")

# Plot actual data
plt.scatter(X, y, color='black', label="Data")

plt.title("Model Complexity vs Fit")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

# 3. Plot training error vs complexity

```python


plt.figure()
plt.plot(degrees, train_errors, marker='o')

plt.title("Training Error vs Model Complexity")
plt.xlabel("Degree")
plt.ylabel("Error")
plt.show()
```

```python
print("Bias:")
print("Error due to overly simple model (underfitting)")

print("\nVariance:")
print("Error due to model being too sensitive to noise (overfitting)")

print("\nOptimal model:")
print("Balance between bias and variance (middle complexity)")
```
