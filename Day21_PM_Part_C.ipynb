### Q1 — Difference between regression and classification

Regression is used when the target variable is continuous. It predicts numerical values.
Example: predicting house prices, temperature, sales.

Classification is used when the target variable is categorical. It predicts classes or labels.
Example: spam detection, fraud detection, disease classification.

Key difference:

* Regression → continuous output
* Classification → discrete/categorical output



### Q2 — calculate_mse(y_true, y_pred)

```python
import numpy as np

def calculate_mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)


# Example
y_true = np.array([3, 5, 7])
y_pred = np.array([2.5, 5, 8])

print(calculate_mse(y_true, y_pred))

```


### Q3 — Bias–Variance Tradeoff

Bias is the error caused by overly simple assumptions in the model.
High bias leads to underfitting.

Variance is the error caused by sensitivity to small changes in training data.
High variance leads to overfitting.

Cases:

* Underfitting → high bias, low variance (model too simple)
* Overfitting → low bias, high variance (model too complex)

Optimal model:

* Achieves balance between bias and variance
* Generalizes well to unseen data
